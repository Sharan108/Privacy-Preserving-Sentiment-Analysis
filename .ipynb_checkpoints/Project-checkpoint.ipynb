```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Reading hospital data
hospital=pd.read_csv("Hospital_Data.csv")
```

```python
#Checking no. rows and columns
hospital.shape
```

```python
#Checking the name of columns
hospital.columns
```

```python
#Partitioning the dataset bases on IRP Value
sensitive=hospital[(hospital['IRP']==0)]
sensitive.shape
non_sensitive=hospital[(hospital['IRP']==1)]
non_sensitive.shape
```

```python
#Removing IRP column from the dataset
columns=['IRP']
sensitive = pd.DataFrame(sensitive.drop(columns,axis=1,inplace=False))
non_sensitive=pd.DataFrame(non_sensitive.drop(columns,axis=1,inplace=False))
```

```python
sensitive.head(n=10)
```

```python
#sensitive['NAME'] = 'Person ' + pd.Series(pd.factorize(sensitive['NAME'])[0] + 1).astype(str)
#sensitive
```

```python
#Retrieving Age from DOB
import datetime as dt
ref_date = dt.datetime.now()
sensitive['DOB'] = sensitive['DOB'].apply(lambda x: len(pd.date_range(start = x, end = ref_date, freq = 'Y')))
non_sensitive['DOB'] = non_sensitive['DOB'].apply(lambda x: len(pd.date_range(start = x, end = ref_date, freq = 'Y')))
sensitive.head
```

```python
sensitive=sensitive.rename(columns={'DOB':'AGE'})
```

```python
# Using Supression technique to preserve name and phone number
sensitive['NAME']='******'
sensitive['PHONE']='**********'
```

```python
#Assigning unique values to the location 
#Performing Enumeration technique
def clean(df,cols):
    for col_name in cols:
        keys={cats:i for i,cats in enumerate(df[col_name].unique())}
        df[col_name]=df[col_name].apply(lambda x:keys[x])
    return df
cols=['ORIGIN']
clean(sensitive,cols)
```

```python
#Applying Binning on DOB
import numpy as np
group_names=["Young","Mid","Old"]
sensitive['AGE']=pd.cut(sensitive['AGE'],bins=[1,25,50,125],labels=group_names,include_lowest=True)
```

```python
#One hot encoding
sensitive['SEX'].replace(['Female','Male'],[0,1],inplace=True)
```

```python
sensitive.head
```

```python
sensitive.head(n=10)
```

```python
df=pd.concat([sensitive,non_sensitive])
```

<h1>Applying Sentiment Analysis</h1> 

```python
import urllib.request
import os
import pandas as pd
import numpy as np
from nltk.tokenize import RegexpTokenizer
from nltk.stem.porter import PorterStemmer
from nltk.corpus import stopwords
```

```python
# Preprocessing of user reviews
tokenizer=RegexpTokenizer(r'\w+')
en_stopwords=set(stopwords.words('english'))
ps=PorterStemmer()
def getStemmedReview(review):
    review=review.lower()
    review=review.replace("<br /><br />"," ")
    #Tokenize
    tokens=tokenizer.tokenize(review)
    new_tokens=[token for token in tokens if token not in  en_stopwords]
    stemmed_tokens=[ps.stem(token) for token in new_tokens]
    clean_review=' '.join(stemmed_tokens)
    return clean_review
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='sentiment',data=df)
temp = df.groupby('sentiment').count()['MESSAGE'].reset_index().sort_values(by='MESSAGE',ascending=False)
temp.style.background_gradient(cmap='Purples')
```

```python
df.drop(['NAME','AGE','PHONE','ORIGIN','SEX','DOCNAME','DISEASE','EXPERT','STAFF','AVAIL','FACILITIES','ENVIRON','FOOD','DURATION'], axis=1)
```

```python
df['MESSAGE'].apply(getStemmedReview)
X_train = df.loc[:121, 'MESSAGE'].values
y_train = df.loc[:121, 'sentiment'].values
X_test = df.loc[121:, 'MESSAGE'].values
y_test = df.loc[121:, 'sentiment'].values
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(sublinear_tf=True, encoding='utf-8',
 decode_error='ignore')
vectorizer.fit(X_train)
X_train=vectorizer.transform(X_train)
X_test=vectorizer.transform(X_test)
```

```python
from sklearn import metrics
from sklearn.metrics import classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Implementing Logistic Regression 
model=LogisticRegression(solver='liblinear')
model.fit(X_train,y_train)
print("Score on training data is: "+str(model.score(X_train,y_train)))
print("Score on testing data is: "+str(model.score(X_test,y_test)))
```

```python
# Implementing Multinomial Naive Bayes Classifier
clf = MultinomialNB()
clf.fit(X_train,y_train)
Y_test_pred = clf.predict(X_test)
print("Score on training data is: "+str(clf.score(X_train,y_train)))
print("Score on testing data is: "+str(clf.score(X_test,y_test)))
```

```python
# Implementing K-Nearest Neighbour
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train,y_train)
pred = knn.predict(X_test)
print("Score on training data is: "+str(knn.score(X_train,y_train)))
print("Score on testing data is: "+str(knn.score(X_test,y_test)))
```

```python
#Conufsion Matrix of the model
cnf_matrix=metrics.confusion_matrix(y_test,pred)
cnf_matrix
```

```python
#Classification Report of the model
print(classification_report(y_test,pred))
print("Accuracy",metrics.accuracy_score(y_test,pred))
```
